```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load mock synthetic data
df = pd.read_csv("synthetic_lacrosse_data.csv")
df['date'] = pd.to_datetime(df['date'])

df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['jump_height_cm'], bins=25, kde=True, color="skyblue")
plt.title("Distribution of Jump Heights - Mock Womens Lacrosse")
plt.xlabel("Jump Height (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='jump_height_cm', y='peak_power_bm', data=df, alpha=0.6)
plt.title("Jump Height vs Peak Power/Body Mass")
plt.xlabel("Jump Height (cm)")
plt.ylabel("Peak Power / Body Mass")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='training_load', data=df, ci=None, estimator='mean')
plt.title("Average Training Load Over Time")
plt.xlabel("Date")
plt.ylabel("Training Load")
plt.xticks(rotation=45)
plt.show()
```
