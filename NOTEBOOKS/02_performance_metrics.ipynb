```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("synthetic_lacrosse_data.csv")
```

```python
df.columns
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    x="jump_height_cm",
    y="rsi",
    data=df,
    alpha=0.6
)
plt.title("Jump Height vs Reactive Strength Index (RSI)")
plt.xlabel("Jump Height (cm)")
plt.ylabel("RSI")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(y="peak_power_bm", data=df)
plt.title("Peak Power / Body Mass Distribution")
plt.ylabel("Peak Power / Body Mass")
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])

plt.figure(figsize=(12,6))
sns.lineplot(
    x="date",
    y="jump_height_cm",
    data=df,
    estimator="mean",
    ci=None
)
plt.title("Average Jump Height Over Time")
plt.xlabel("Date")
plt.ylabel("Jump Height (cm)")
plt.xticks(rotation=45)
plt.show()
```

```python
### Key Performance Insights
- Jump height shows stable trends across sessions with moderate variability.
- RSI demonstrates a positive relationship with jump height, indicating neuromuscular efficiency.
- Peak power distribution suggests consistency across the squad, with identifiable high performers.
```
